```python
import os
import shutil
from pathlib import Path

# 1. Automatically find the actual path to sequences and annotations
def find_dir(name, path):
    for root, dirs, files in os.walk(path):
        if name in dirs:
            return os.path.join(root, name)
    return None

# Use the 'path' variable from Cell 1
real_images_path = find_dir("sequences", path)
real_labels_path = find_dir("annotations", path)

if not real_images_path or not real_labels_path:
    print("❌ Error: Could not find 'sequences' or 'annotations' in the downloaded files.")
    print(f"Contents of download directory: {os.listdir(path)}")
else:
    print(f"✅ Found images at: {real_images_path}")
    print(f"✅ Found labels at: {real_labels_path}")

    # 2. Define target paths
    base_dir = Path("/content/yolo_dataset")
    images_dst = base_dir / "images/train"
    labels_dst = base_dir / "labels/train"

    images_dst.mkdir(parents=True, exist_ok=True)
    labels_dst.mkdir(parents=True, exist_ok=True)

    # 3. Restructure
    print("Restructuring files... please wait.")
    src_images = Path(real_images_path)
    src_labels = Path(real_labels_path)

    for sequence in os.listdir(src_images):
        seq_img_path = src_images / sequence
        seq_lbl_path = src_labels / sequence

        if seq_img_path.is_dir():
            for img_file in os.listdir(seq_img_path):
                if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
                    # Copy Image
                    shutil.copy(seq_img_path / img_file, images_dst / f"{sequence}_{img_file}")

                    # Copy Label
                    lbl_file = img_file.rsplit('.', 1)[0] + ".txt"
                    if (seq_lbl_path / lbl_file).exists():
                        shutil.copy(seq_lbl_path / lbl_file, labels_dst / f"{sequence}_{lbl_file}")

    print(f"🚀 Success! Images: {len(os.listdir(images_dst))}, Labels: {len(os.listdir(labels_dst))}")
```
